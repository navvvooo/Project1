```python
df = pd.read_csv(r'C:\Users\User\Downloads\DDW-0000C-24.csv') #import csv file
#df.columns
```

```python
df1=df[[ 'Unnamed: 3','Unnamed: 1','Unnamed: 4','C-24 DISABLED NON-WORKERS BY TYPE OF DISABILITY AND SEX', 'Unnamed: 6']]
data =df1.copy() #copying col
#data.head(20)
```

```python
data.loc[data['Unnamed: 4']=='Total'] #filter
```

```python
f =data.iloc[30:, :] #removing useless rows
#f.head(20)
```

```python
data2 = f.loc[f['C-24 DISABLED NON-WORKERS BY TYPE OF DISABILITY AND SEX'] =='Total'] #filter
```

```python
g= data2.sort_values(by=['Unnamed: 4','Unnamed: 1']) #sort
```

```python
g['Unnamed: 6']=g['Unnamed: 6'].astype('int') #type convert
```

```python
g1= g.reset_index(drop=True) #reindexing
```

```python
n1=g1.iloc[0:35] #rural
n2= g1.iloc[35:70] #total
n3=g1.iloc[71:105] #urban
a=n1.plot(x='Unnamed: 1',y='Unnamed: 6',figsize=(10,5),color='red')
b=n2.plot(x='Unnamed: 1',y='Unnamed: 6',figsize=(10,5))
c=n3.plot(x='Unnamed: 1',y='Unnamed: 6',figsize=(10,5),color='green')
```

```python
ax = n1.plot("Unnamed: 1", "Unnamed: 6", kind="line", label = 'Urban')
n3.plot("Unnamed: 1", "Unnamed: 6", ax= ax , kind="line", label = 'Rural', title= 'MultiLine Plot',figsize=(10,5)) #combineplot
```
